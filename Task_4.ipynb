```python
import numpy as np
from scipy.integrate import odeint
import pylab as plt
```

```python
def equation(y, t):
    v, w = y
    dydt = [v + w, v**2 - w**2]
    return dydt
```

```python
y0 = [1., 2.]
t1 = 2
t2 = 3
t = np.arange(2, 3, 0.05)
sol = odeint(equation, y0, t)
v, w = sol[:, 0], sol[:, 1]
```

```python
plt.plot(t, v, 'g', label = 'v (t)')
plt.plot(t, w, 'r', label = 'w (t)')
plt.legend(loc = 'best')
plt.xlabel('t')
plt.grid()
plt.ylim([-0,10])
plt.show()
```
