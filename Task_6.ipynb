## Task 11.9.2 (c)

Solve a nonlinear boundary value problem::  
$y'' + sin(y) = 0, 0 < x < x_k,$  
$y(0) = 0, y(x_k) = \pi,$   
$x_k = 0.5, 1, 2, 4, 6$

```python
import numpy as np
from scipy.integrate import solve_bvp
import pylab as plt
```

```python
def equation(t, y):
    return np.vstack((y[1], -np.sin(y[0])))
```

```python
def bc(ya, yb):
    return np.array([ya[0], yb[0] - np.pi])
```

```python
def solver(bv_0, bv_1, step):
    t = np.arange(bv_0, bv_1, step)
    y = np.zeros((2, t.size))
    return solve_bvp(equation, bc, t, y)
```

```python
def plotter(bv_from, bv_to, solution, j):
    plt.subplot(2, 3, j)
    t_array = np.arange(bv_from, bv_to, 0.001)  
    plt.plot(t_array, solution.sol(t_array)[0], label='y')
    plt.legend(loc = 'best')
    plt.grid()
    plt.xlabel("t")
    plt.ylabel("y")
```

```python
def solve_equation(step):
    plt.figure(figsize=(18, 12))
    for index, x in enumerate([0.5, 1, 2, 4, 6], 1):
        plotter(0, x, solver(0, x, step), index)
```

```python
solve_equation(step=1e-1)
```

```python
solve_equation(step=1e-2)
```

```python
solve_equation(step=1e-4)
```
